# Pipeline Feature Correlation Report

Checks which default `ForexPipeline` features are highly correlated for a selected pair/year/timeframe. Default config matches the current XGBoost-style setup: EURUSD, 2023, H1, fracdiff d=0.3.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
for _ in range(6):
    if (ROOT / "histdata").is_dir() and (ROOT / "Pipeline").is_dir():
        break
    ROOT = ROOT.parent
else:
    raise RuntimeError("Run this notebook from inside the Trade_bot repository")

sys.path.insert(0, str(ROOT))
from Supplementary.feature_correlation import run_report, print_summary
```

## Config

```python
PAIR = "EURUSD"
YEARS = [2023]
TIMEFRAME = "H1"
WEEKENDS = "filled"
NORM_METHOD = "fracdiff"
FRACDIFF_D = 0.3
THRESHOLD = 6e-4
CORR_METHOD = "pearson"
HIGH_CORR_THRESHOLD = 0.85
OUT_DIR = ROOT / "Supplementary" / "feature_correlation_outputs
```

## Run Report

```python
report = run_report(
    pair=PAIR,
    years=YEARS,
    timeframe=TIMEFRAME,
    weekends=WEEKENDS,
    norm_method=NORM_METHOD,
    fracdiff_d=FRACDIFF_D,
    threshold=THRESHOLD,
    corr_method=CORR_METHOD,
    high_corr_threshold=HIGH_CORR_THRESHOLD,
    out_dir=OUT_DIR,
    save=True,
)
print_summary(report)
```
